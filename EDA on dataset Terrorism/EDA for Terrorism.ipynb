```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('terrorism.csv', encoding='latin1')
```

```python
data
```

```python
# Displaying the first few rows of the dataset
print(data.head())

# Checking the shape of the dataset
print(data.shape)

# Getting an overview of the columns and their data types
print(data.info())

# Checking for missing values
print(data.isnull().sum())
```

```python
# Plotting a histogram of the number of terrorist incidents per year
plt.figure(figsize=(12, 6))
sns.histplot(data['iyear'], kde=False, bins=30)
plt.title('Number of Terrorist Incidents per Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
# Finding the top 10 countries with the most terrorist incidents
top_countries = data['country_txt'].value_counts().head(10)
print(top_countries)

# Plotting a bar chart of the top 10 countries with the most terrorist incidents
plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.index, y=top_countries.values)
plt.title('Top 10 Countries with the Most Terrorist Incidents')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
