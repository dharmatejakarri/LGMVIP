```python
import cv2

# Read the image
image = cv2.imread('Cat.jpg')

# Convert the image to RGB format
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
image
```

```python
# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
```

```python
# Invert the grayscale image
inverted_image = cv2.bitwise_not(gray_image)
```

```python
# Create the pencil sketch by dividing grayscale image by inverted blurry image
blurred_image = cv2.GaussianBlur(inverted_image, (111, 111), 0)
pencil_sketch = cv2.divide(gray_image, blurred_image, scale=256.0)
```

```python
import matplotlib.pyplot as plt

# Display the pencil sketch
plt.imshow(pencil_sketch, cmap='gray')
plt.axis('off')
plt.show()
```
